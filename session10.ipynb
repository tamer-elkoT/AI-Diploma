Execice 1 >> Extract quotes

```python
#define the webage  url
url = "https://quotes.toscrape.com/"

#send a get request to the server that contain the html code
response = requests.get(url)
print(response) # o/p >> <Response [200]> means OK>>
#The output <Response [200]> indicates that the HTTP GET request to the server was successful
```

```python
# To get the content of the  page (HTMl code) 
print(response.content)
```

```python
print(response.text)
```

```python
# parse the HTML content of the response using Beutifulsoup
# what is the meaning of parsing ?
# parsing is the process of breaking down a string of text into a collection of words or tokens,
# which can then be analyzed or processed further.
#For example, when you write code, the compiler or interpreter parses your code to understand its
# structure and syntax before converting it into machine code that the computer can execute
soup = BeautifulSoup(response.content,"html.parser") #note that we have two parsers XML and HTML

print(soup)


# you note that the code is more readable after using beautifulsoap
```

```python
print(type(soup))
```

```python
# if you go to the webpage you note that all quotes are into span tag its class name is text after inspect
# so we will use this class name to find all the paragraph
# now soup is the object
#  that i have created as a tool of beautifulsoup library
# so i can use soup in various things like get title , head ,parent,child,find,find_all() etc....

# title_tag = soup.title
# print(title_tag) #Quotes to Scrape
# print(soup.get_text())
quotes = soup.find_all("span",class_="text")
print(f"Number of spam tags : {len(quotes)}")

print(quotes)
# we can index the tags as a list item
print(quotes[0])

```

```python
# we can iterate over the list of tags because each item in the list contain one span
# so i can truncate the text only by using >> get_text()

for quote in quotes:
    # get the text of the quote
    text = quote.get_text()
    print(text)
```

```python
#find the next button 
next_button = soup.find("li",class_="next")
print(next_button)
print("__________________________________________")
print(next_button.a)
print("__________________________________________")
print(next_button.a["href"])
print("__________________________________________")
print(f"New URL : {url + next_button.a["href"]}")
```

Application >> Download an image from a website into file in your directory 

```python
# find the next button 
next_button = soup.find("li",class_= "next")
# while the next button is exist >> there is an exist page when i click the next button

while next_button:
    #get the url of the next page
    next_url = url + next_button.a["href"]

    # send a get request to the next page
    response = requests.get(next_url)


    # pase the HTML content
    soup = BeautifulSoup(response.content,"html.parser")

    # find all quotes on the page 
    quotes = soup.find_all("span",class_= "text")

    #print each quote 
    for quote in quotes :
        print(quote.get_text())
    #find the next button of the current page 
    next_button = soup.find("li",class_= "next")
```

```python
#get the url of the website
base_url = "https://www.crummy.com/software/BeautifulSoup/bs4/doc/"
# get the html content of the website
response2 = requests.get(base_url)

soup2 = BeautifulSoup(response2.content,"html.parser")
#find the image tag after inspect 
img = soup2.find("img",class_="align-right")

#to get the absolute url of the image 
from urllib.parse import urljoin

img_url = urljoin(base_url, img['src']) 

#print(img['src']) #_images/6.1.jpg >> note that this line gave me the relative url not the absolute url of the image
#and it will raise error if i use the relative url

img_data= requests.get(img_url).content
print(img_data) # it print the binary code of the image

# using file handling i open a new file named Downloaded file
# save the file as an object in a varible called handler
# and write the binary code of the image in it

with open("Downloaded_image.jpg", "wb") as handler:
    handler.write(img_data)

print("image downloaded successfully")

```

Book information


```python
url = "https://books.toscrape.com/"

#send an HTTP get request to the website
response = requests.get(url)

#parse the HTMLL content using Beautiful soup
soup = BeautifulSoup(response.content,"html.parser")
```

```python
#find all book containers 
book_container = soup.find_all("article",class_= "product_pod")
#print(book_container[0]) it print the info of the first book

# Iterate through each book container and extract information 
for book in book_container:
    #print(book) # the book give the info of each book in every iteration
    # Extract the title of the book
    title = book.h3.a["title"]
    # print(title)

    #Extact book price 
    # price = book.find("p",class_= "price_color") #  it print the whole info of the paragraph p put i need only the text price
    price = book.find("p",class_= "price_color").get_text()
    # print(price)

    #Extact star rating
    rating = book.find('p',class_= 'star-rating' )['class'][-1]
    # print(rating)

    print(f"Book Title is : {title}")
    print(f"Book Price is : {price}")
    print(f"Book rating is : {rating}")


```
